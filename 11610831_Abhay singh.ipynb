```python
import matplotlib.pyplot as plt

population_india_census2011['Gender Ratio'].plot(kind='pie',
                            figsize=(19,8),
                            autopct='%1.1f%%', 
                            startangle=60,  
                            pctdistance=1.12,
                            labels=None,   
                            )

plt.title('Total Positive Cases according to State')
plt.legend(labels=population_india_census2011.index, loc='upper left') 
plt.axis('equal')

plt.show()
```
